```python
#WEB SCRAPING WITH TOPICS LINK

#w3 schools python index
import requests
from bs4 import BeautifulSoup

def get_w3schools_titles():
    url = 'https://www.w3schools.com/python/default.asp'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    menu = soup.find('div', id='leftmenuinnerinner')
    links = menu.find_all('a')
    
    titles = [link.text.strip() for link in links if link.text.strip()]
    return set(titles)
    
#tpoint tech python index
def get_tutorialspoint_titles():
    url = 'https://www.tutorialspoint.com/python/index.htm'
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')

    menu = soup.find('ul', class_='toc chapters')
    links = menu.find_all('a')
    
    titles = [link.text.strip() for link in links if link.text.strip()]
    return set(titles)
    
# Get data
w3_titles = get_w3schools_titles()
tp_titles = get_tutorialspoint_titles()


# Compare
common = w3_titles & tp_titles
only_w3 = w3_titles - tp_titles
only_tp = tp_titles - w3_titles

# Print results
print("üî• Common Topics:")
for title in sorted(common):
    print("  -", title)

print("\nüåê Unique to W3Schools:")
for title in sorted(only_w3):
    print("  -", title)

print("\nüìò Unique to TutorialsPoint:")
for title in sorted(only_tp):
    print("  -", title)
```
